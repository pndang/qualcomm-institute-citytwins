```python
import csv
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
def geocode_address(address):
    geolocator = Nominatim(user_agent="geocoder_app")
    location = geolocator.geocode(address)

    if location:
        latitude = location.latitude
        longitude = location.longitude 
        return latitude, longitude 
    else:
        return None, None 
```

```python
# import data

df = pd.read_csv("costar-affordable-housing-data.csv", index_col=0)
df.head()
```

```python
# List of addresses with names
# addresses_with_names = [
#     {"name": "Atkinson Hall", "address": "3195 Voigt Dr, La Jolla, CA 92093"},
#     {"name": "Canyonview Aquatic Center", "address": "3302 Voigt Dr, San Diego, CA 92121"}
# ]

addresses_with_names = []

def fillAddressStruct(row):
    propName = row['Building Name']
    propAddress = row['Address']
    addresses_with_names.append({"name": propName, "address": propAddress})
    return None 

df.apply(fillAddressStruct, axis=1);
```

```python
# Geocode each address above and save to CSV
idx = 1
# with open("geocoded_addresses.csv", mode='w', newline='') as 
with open("geocoded_costar_addresses.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow([0, 'Name', 'Address', 'Latitude', 'Longitude'])

    for item in addresses_with_names:
        name = item['name']
        address = item['address']
        address = address + ', San Diego'

        lat, lon = geocode_address(address)

        writer.writerow([idx, name, address, lat, lon])

        idx += 1

# print("Geocoded addresses saved to 'geocoded_addresses.csv'.")
print("Saved!")
```
